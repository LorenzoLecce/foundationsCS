## Solutions

````python
import re
````

**1)**

````python
numbers = open('numbers.txt', 'r').readlines()
even = 0
even_re = re.compile('^\d*[02468]\n')
for line in numbers:
    occ = even_re.search(line)
    if occ:
        even += 1
print(even)
````

````python
numbers = open('numbers.txt', 'r').read()
even = re.findall('\d*[02468]\n', numbers)
print(len(even))
````

**2)**

````python
emails = open('email.txt', 'r').readlines()
re_email = re.compile('\s*(?P<name>\w+)[\._!](?P<surname>[a-z]+)(?P<number>\d*)@(?P<domain>\S+)\s+(?P<age>\d+)', re.I)
# re_email = re.compile('\s*(?P<name>\w+)[\._!](?P<surname>[a-z]+)(?P<number>\d*)@(?P<domain>(\w+\.?)+)\s+(?P<age>\d+)', re.I)
for line in emails[:1000]:
    occ = re_email.search(line)
    if occ:
        print(line)
        print('Name:', occ.group('name'))
        print('Surname:', occ.group('surname'))
        print('Number:', occ.group('number'))
        print('Domain:', occ.group('domain'))
        print('Age:', occ.group('age'))
        print('-'*80)
        
    # TODO: count the domains
````

**3)**

````python
expnumbers = open('exp_nums.txt', 'r').readlines()
re_enum = re.compile('^(?P<number>[0-9]\.\d+)e(?P<exp>\d+)')
for line in expnumbers:
    occ = re_enum.search(line)
    if occ:
        print('Number:', occ.group('number'))
        print('Exponential: ', occ.group('exp'))
        print('Conversion: ', float(occ.group('number')) * 10**(int(occ.group('exp'))))
````

**4)**

````python
def camel_to_snake(camelString):
    occ = re.findall('[A-Z][a-z]+', camelString)
    words = []
    for word in occ:
        words.append(word.lower())
    return '_'.join(words)

print(camel_to_snake('CamelCamelCamel'))
````

````python
# this is more complicated
def camel_to_snake(camelString):
    occ = re.findall('(.)([A-Z])', camelString)
    print(occ)
    return re.sub(r'(.)([A-Z])', r'\1_\2', camelString).lower()

print(camel_to_snake('CamelSnakeTiger'))
````
